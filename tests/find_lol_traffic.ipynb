```python
from bat import bro_log_reader
from pprint import pprint
```

```python
reader = bro_log_reader.BroLogReader('../test_2/ssl.log')
```

```python
for row in reader.readrows():
    if row['id.orig_p'] >= 4900:
        print(row)
```

```python
ports = []
for row in reader.readrows():
    if ('riot' or 'lol') in (row['server_name'] or row['subject']):
        ports.append({'host':row['id.orig_h'],'port':row['id.orig_p']})
```

```python
print(ports)
```

```python
ordered = sorted(ports, key = lambda k: k['port'])
print(ordered[0])
print(ordered[-1])
```
